```python
intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey"],
        "responses": ["Hello!", "Hi there!", "Hey!"]
    },
    "price_query": {
        "patterns": ["price", "cost", "rate"],
        "responses": ["Please tell me the commodity name."]
    },
    "goodbye": {
        "patterns": ["bye", "exit"],
        "responses": ["Goodbye!", "See you later!"]
    }
}

commodities = {
    "potato": 30,
    "tomato": 50,
    "onion": 40,
    "rice": 60,
    "wheat": 35,
    "sugar": 45
}
```

```python
import random

def get_intent(user_input):
    user_input = user_input.lower()
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return intent
    return "unknown"

def get_commodity_price(user_input):
    user_input = user_input.lower()
    for commodity, price in commodities.items():
        if commodity in user_input:
            return commodity, price
    return None, None

def chatbot():
    waiting_for_commodity = False

    while True:
        user_input = input("User: ")
        intent = get_intent(user_input)

        # If we were waiting for a commodity name, check for it first
        if waiting_for_commodity:
            commodity, price = get_commodity_price(user_input)
            if commodity:
                print(f"Bot: The price of {commodity} is Rs.{price}/kg.")
                waiting_for_commodity = False
                continue
            else:
                print("Bot: Sorry, I don't have the price for that commodity.")
                waiting_for_commodity = False
                continue

        # Check if the user directly asks price of a specific commodity
        if intent == "price_query":
            commodity, price = get_commodity_price(user_input)
            if commodity:
                print(f"Bot: The price of {commodity} is Rs.{price}/kg.")
            else:
                response = random.choice(intents[intent]["responses"])
                print("Bot:", response)
                waiting_for_commodity = True

        elif intent == "unknown":
            print("Bot: I didn't understand that.")
        else:
            response = random.choice(intents[intent]["responses"])
            print("Bot:", response)

        if intent == "goodbye":
            break

chatbot()
```
