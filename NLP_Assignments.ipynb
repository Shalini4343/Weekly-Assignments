```python
# Assignment 1 — Text Preprocessing

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng')


from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords,wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
# Step 1: Input Text (2 Paragraphs with Multiple Punctuation Marks)

text = """
Artificial Intelligence, often abbreviated as AI, has transformed numerous industries: healthcare,
finance, education, and transportation! Researchers around the world are asking — can machines truly
"think" like humans? While some experts believe it's possible; others remain skeptical. The key
technologies driving AI include: neural networks, deep learning, and natural language processing (NLP).
As of 2025, AI-powered tools are everywhere — from chatbots to self-driving cars!

Natural Language Processing (NLP) is a fascinating sub-field of AI; it focuses on the interaction
between computers and human language. Tasks such as: sentiment analysis, machine translation, and
text summarization are now commonplace! However, challenges remain — ambiguity, sarcasm, and
context-dependent meanings make NLP incredibly difficult. For instance, the sentence "I saw her duck"
has multiple interpretations; is it a noun or a verb? Researchers continue to push boundaries,
developing newer models every year.
"""

print("=== Original Text ===")
print(text)
```

```python
# Step 2: Tokenization

# Tokenize into words
tokens = word_tokenize(text)

print("Total tokens: ",len(tokens))
print("First 30 tokens: ",tokens[:30])
```

```python
# Step 3: Removing Punctuation Marks

# Remove punctuation keep only alphabetic tokens
tokens_no_punct = [token for token in tokens if token.isalpha()]

# Display
print("Tokens after punctuation removal: ",len(tokens_no_punct))
print(tokens_no_punct[:30])
```

```python
#Step 4: Stop Word Removal

stop_words = set(stopwords.words('english'))
tokens_no_stop = [token for token in tokens_no_punct if token.lower() not in stop_words]

# Display

print("Tokens after stop word removal: ",len(tokens_no_stop))
print(tokens_no_stop)
```

```python
# Step 5: Stemming (Porter Stemmer)

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in tokens_no_stop]

print("All stemmed tokens: ",len(stemmed_tokens))
print(stemmed_tokens)
```

```python
# Step 6: Lemmatization (WordNet Lemmatizer)

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens_no_stop]

print("All lemmatized tokens:",len(lemmatized_tokens))
print(lemmatized_tokens)
```

```python
 # Step 7: Lemmatization with POS Tagging (WordNet Lemmatizer)
lemmatizer1 = WordNetLemmatizer()

def get_pos(word):
    tag = pos_tag([word])[0][1][0].upper()

    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }

    return tag_dict.get(tag, wordnet.NOUN)

lemmatized_tokens1 = [lemmatizer1.lemmatize(token, get_pos(token)) for token in tokens_no_stop]

print("Lemmatized tokens: ",len(lemmatized_tokens1))
print(lemmatized_tokens1)
```

```python
# Assignment 2 — Feature Engineering & Word2Vec

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
import pandas as pd
```

```python
# Part 1: Bag of Words (BoW)
converted_text = " ".join(lemmatized_tokens1)
print(converted_text)
```

```python
bow_words = [converted_text]
bow_vectorizer = CountVectorizer()
bow_vectorizer_words = bow_vectorizer.fit_transform(bow_words)

print(bow_vectorizer.get_feature_names_out())
print(bow_vectorizer_words.toarray())
```

```python
# TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_words = [converted_text]
tfidf_vectorizer_words = tfidf_vectorizer.fit_transform(tfidf_words)

print(tfidf_vectorizer.get_feature_names_out())
print(tfidf_vectorizer_words.toarray())
```

```python
w2v_text = """
The history of computing stretches back thousands of years, from the ancient abacus to modern
quantum computers. Charles Babbage is often called the father of computing for designing the
Analytical Engine in the 1830s, a mechanical general-purpose computer. Ada Lovelace, working
alongside Babbage, wrote what is considered the first computer program. The twentieth century
saw explosive growth in computing technology. Alan Turing laid the theoretical foundations with
his concept of the Turing machine, which formalized the notion of computation and algorithms.
During World War II, machines like Colossus and ENIAC were built to perform complex calculations
at unprecedented speeds. The invention of the transistor in 1947 revolutionized electronics,
leading to smaller, faster, and more reliable computers. The integrated circuit followed,
packing thousands of transistors onto a single chip. The personal computer revolution of the
1980s brought computing power to homes and offices worldwide. Companies like Apple, IBM, and
Microsoft became household names. The internet, originally a military research project called
ARPANET, connected millions of computers across the globe, transforming communication, commerce,
and entertainment. Today, artificial intelligence and machine learning represent the latest
frontier in computing. Deep learning models process vast amounts of data to recognize patterns,
generate text, translate languages, and even create art. Cloud computing allows businesses to
scale infrastructure on demand, while edge computing brings processing closer to data sources.
Quantum computing promises to solve problems that are currently intractable for classical
computers, potentially revolutionizing fields like cryptography, drug discovery, and climate
modeling. The journey from the abacus to quantum processors illustrates humanity's relentless
drive to compute faster, think bigger, and solve harder problems.
"""

print("Word count: ",len(w2v_text.split()))
```

```python
# tokenize into list of sentences then list of words and punctuation removal

w2v_sentences_raw = sent_tokenize(w2v_text.strip())
w2v_sentences = []
for sent in w2v_sentences_raw:
    tokens = word_tokenize(sent)
    tokens = [t.lower() for t in tokens if t.isalpha()]
    w2v_sentences.append(tokens)

print("Number of sentences: ",len(w2v_sentences))
print(w2v_sentences)
```

```python
# Train Word2Vec model
model = Word2Vec(
    w2v_sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    #epochs=100,
)
```

```python
# Check similarity

print("Vector for microsoft")
print(model.wv['microsoft'])

print("words similar to microsoft")
print(model.wv.most_similar('microsoft'))
```

```python
print("Vector for apple")
print(model.wv['apple'])

print("words similar to apple")
print(model.wv.most_similar('apple'))
```

```python
# Check similarity scores between word pairs

score = model.wv.similarity("computing","computers")
print("Word Pair Similarity Scores")
print("computing  ->","computers    :",score)
```

```python
score = model.wv.similarity("machine","learning")
print("Word Pair Similarity Scores")
print("machine  ->","learning    :",score)
```
